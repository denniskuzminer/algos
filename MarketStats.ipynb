#### Get derived columns

```python
MARKET_OPEN = "09:30:00"
MARKET_CLOSE = "16:00:00"
# Market Pulse – 8 EMA > 21 EMA and 21 EMA > 34 EMA
computed_data["Market Pulse"] = ((computed_data["8 EMA"] > computed_data["21 EMA"]) & (computed_data["21 EMA"] > computed_data["34 EMA"])) # .apply(lambda x: "Yes" if x else "No")
# VWAP... ASK FOR ANCHOR DATE
result = indicators.get_vwap(quotes_list)
vwap_df = pd.DataFrame.from_dict({
    "DateTime": [r.date for r in result],
    "VWAP": [r.vwap for r in result],
})
computed_data = computed_data.merge(vwap_df, how="inner", on="DateTime")
# Close Price >= VWAP (Yes or No)
computed_data["Close Price >= VWAP"] = (computed_data["Close"] >= computed_data["VWAP"]) # .apply(lambda x: "Yes" if x else "No")
# Close Price >= 8 EMA (Yes or No)
computed_data["Close Price >= 8 EMA"] = (computed_data["Close"] >= pd.to_numeric(computed_data["8 EMA"], errors='coerce')) # .apply(lambda x: "Yes" if x else "No")
# Close Price >= 13 EMA (Yes or No)
computed_data["Close Price >= 13 EMA"] = (computed_data["Close"] >= pd.to_numeric(computed_data["13 EMA"], errors='coerce')) # .apply(lambda x: "Yes" if x else "No")
# Close Price >= 21 EMA (Yes or No)
computed_data["Close Price >= 21 EMA"] = (computed_data["Close"] >= pd.to_numeric(computed_data["21 EMA"], errors='coerce')) # .apply(lambda x: "Yes" if x else "No")
# Close Price >= 34 EMA (Yes or No)
computed_data["Close Price >= 34 EMA"] = (computed_data["Close"] >= pd.to_numeric(computed_data["34 EMA"], errors='coerce')) # .apply(lambda x: "Yes" if x else "No")
# Close Price >= 50 EMA (Yes or No)
computed_data["Close Price >= 50 EMA"] = (computed_data["Close"] >= pd.to_numeric(computed_data["50 EMA"], errors='coerce')) # .apply(lambda x: "Yes" if x else "No")
# Close Price >= 200 EMA (Yes or No)
computed_data["Close Price >= 200 EMA"] = (computed_data["Close"] >= pd.to_numeric(computed_data["200 EMA"], errors='coerce')) # .apply(lambda x: "Yes" if x else "No")
# 21 EMA >= 34 EMA (Yes or No)
computed_data["21 EMA >= 34 EMA"] = (pd.to_numeric(computed_data["21 EMA"], errors='coerce') >= pd.to_numeric(computed_data["34 EMA"], errors='coerce')) # .apply(lambda x: "Yes" if x else "No")
# 21 EMA >= 50 EMA (Yes or No)
computed_data["21 EMA >= 50 EMA"] = (pd.to_numeric(computed_data["21 EMA"], errors='coerce') >= pd.to_numeric(computed_data["50 EMA"], errors='coerce')) # .apply(lambda x: "Yes" if x else "No")
# 21 EMA >= 200 EMA (Yes or No)
computed_data["21 EMA >= 200 EMA"] = (pd.to_numeric(computed_data["21 EMA"], errors='coerce') >= pd.to_numeric(computed_data["200 EMA"], errors='coerce')) # .apply(lambda x: "Yes" if x else "No")
# Regular market hours
computed_data["Regular Market Hours"] = (computed_data["DateTime"].dt.time >= pd.to_datetime(MARKET_OPEN).time()) & (computed_data["DateTime"].dt.time <= pd.to_datetime(MARKET_CLOSE).time())
computed_data.to_csv("./MarketStats.csv")
computed_data = computed_data.infer_objects()
computed_data
```

#### Add Technical Indicators

```python
# MACD... ASK FOR FAST, SLOW, AND SIGNAL PERIODS
result = indicators.get_macd(quotes_list)
macd_df = pd.DataFrame.from_dict({
    "DateTime": [r.date for r in result],
    "MACD": [r.macd for r in result],
    "MACD Fast EMA": [r.fast_ema for r in result],
    "MACD Slow EMA": [r.slow_ema for r in result],
})
computed_data = computed_data.merge(macd_df, how="inner", on="DateTime")
# CCI... ASK FOR LOOKBACK PERIODS
result = indicators.get_cci(quotes_list)
cci_df = pd.DataFrame.from_dict({
    "DateTime": [r.date for r in result],
    "CCI": [r.cci for r in result],
})
computed_data = computed_data.merge(cci_df, how="inner", on="DateTime")
# ATR... ASK FOR LOOKBACK PERIODS
result = indicators.get_atr(quotes_list)
atr_df = pd.DataFrame.from_dict({
    "DateTime": [r.date for r in result],
    "ATR": [r.atr for r in result],
})
computed_data = computed_data.merge(atr_df, how="inner", on="DateTime")
# ZScore... ASK FOR LOOKBACK PERIODS
result = indicators.get_stdev(quotes_list, 120)
stdev_df = pd.DataFrame.from_dict({
    "DateTime": [r.date for r in result],
    "ZScore": [r.stdev for r in result],
})
computed_data = computed_data.merge(stdev_df, how="inner", on="DateTime")
# Parabolic SAR... ASK FOR ACCELERATION STEP AND MAX ACCELERATION FACTOR
result = indicators.get_parabolic_sar(quotes_list)
sar_df = pd.DataFrame.from_dict({
    "DateTime": [r.date for r in result],
    "Parabolic SAR": [r.sar for r in result],
})
computed_data = computed_data.merge(sar_df, how="inner", on="DateTime")
computed_data["Parabolic SAR >= Close"] = (computed_data["Parabolic SAR"] >= computed_data["Close"])
# TTM Squeeze - https://github.com/hackingthemarkets/ttm-squeeze
```

```python
computed_data['20 SMA'] = computed_data['Close'].rolling(window=20).mean()
computed_data['Lower Bollinger Band'] = pd.to_numeric(computed_data['20 SMA'], errors='coerce') - (2 * pd.to_numeric(computed_data['ZScore'], errors='coerce'))
computed_data['Upper Bollinger Band'] = pd.to_numeric(computed_data['20 SMA'], errors='coerce') + (2 * pd.to_numeric(computed_data['ZScore'], errors='coerce'))
computed_data['Lower Keltner Channel'] = pd.to_numeric(computed_data['20 SMA'], errors='coerce') - (pd.to_numeric(computed_data['ATR'], errors='coerce') * 1.5)
computed_data['Upper Keltner Channel'] = pd.to_numeric(computed_data['20 SMA'], errors='coerce') + (pd.to_numeric(computed_data['ATR'], errors='coerce') * 1.5)
computed_data['TTM Squeeze On'] = ((computed_data['Lower Bollinger Band'] > computed_data['Lower Keltner Channel']) & (computed_data['Upper Bollinger Band'] < computed_data['Upper Keltner Channel'])) # .apply(lambda x: "Yes" if x else "No")
computed_data["Squeeze Fired"] = (computed_data['TTM Squeeze On'].shift() & ~computed_data['TTM Squeeze On'])
# computed_data[computed_data["Squeeze Fired"]]
```
